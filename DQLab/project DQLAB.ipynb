# Outliers

```python
import numpy as np
import pandas as pd

Q1 = df_listing[['price']].quantile(0.25)
Q1
```

```python
Q3 = df_listing[['price']].quantile(0.75)
Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_IQR = Q1 - 1.5 * IQR
lower_IQR
```

```python
upper_IQR = Q3 + 1.5 * IQR
upper_IQR
```

```python
df_listing.boxplot(column='price')
```

```python
import seaborn as sns

sns.boxplot(df_listing['price'])
```

```python
print('before removing outliers (shape) : ',df_listing.shape)
df_listing_outliers = df_listing[~((df_listing < lower_IQR) |(df_listing > upper_IQR)).any(axis=1)]
print('after removing outliers (shape) : ',df_listing_outliers.shape)
print('---------------------------------')
print('before removing outliers : ', len(df_listing))
print('after removing outliers : ', len(df_listing_outliers))
print('outliers : ', len(df_listing)-len(df_listing_outliers))
```
